```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/Data Sets/Petrol & Diesel Price (2017-2024).csv')
df.head()
```

```python
# Group by 'Month' and 'Year' and calculate the average of 'Delhi(P)'

grouped = df[df['Year'] == 2018]

grouped = grouped.groupby(['Month', 'Year']).agg({'Delhi(P)': 'mean'}).reset_index()

# Sort the DataFrame by 'Month'
grouped = grouped.sort_values('Month')

# Create a pivot table for plotting
pivot_table = grouped.pivot(index='Month', columns='Year', values='Delhi(P)')


# Plot the data
pivot_table.plot.bar(figsize = (10,5),grid = True)

tick = range(len(pivot_table))
plt.xticks(ticks = tick, labels = pivot_table.index, rotation = 45)

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.xlabel('Month')
plt.ylabel('Average of Petrol Price in Delhi')
plt.title('Average of Petrol Price in Delhi by Month and Year')
plt.show()
```
